```python
% matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('freq.xlsx')
df
```

```python
# This is a figure to show the freq of different defective chr behavior in control/ exp PtK2 cells.
fig = plt.figure(figsize=(10, 4))
plt.title('Frequency Analysis')
sns.heatmap(df, cmap ='coolwarm')
```

```python
# This is a figure to show the freq of different defective chr behavior in control/ exp PtK2 cells with annotations.
fig = plt.figure(figsize=(10, 4))
plt.title('Frequency Analysis')
sns.heatmap(df, cmap ='coolwarm', annot= df)
```

```python
# Let's try the clustering algo to see cells behavior similiarity.
sns.clustermap(df, cmap ='coolwarm', col_cluster=False, figsize=(10, 5))
```
